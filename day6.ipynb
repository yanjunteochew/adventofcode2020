```python
import pandas as pd
import string
```

```python
lists = [[]]
for x in open('data/input-day6.txt').read().split('\n\n'):
    l = x.split()
    lists.append(l)
lists[0]
```

```python
total = 0
for l in lists:
    passengergroup = ''
    for passenger in l:
        passengergroup = passengergroup + passenger
    s = set(passengergroup)
    total = total + len(s)
```

```python
total
```

```python
total = 0
for l in lists:
    print(l)
    passengerset = set(string.ascii_lowercase)
    if not l:
        continue
    for passenger in l:
        passengerset = set.intersection(passengerset, set(passenger))
    total = total + len(passengerset)
```
