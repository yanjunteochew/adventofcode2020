```python
def updateacc(value, acc):
    op = value[:1]
    val = value[1:]
    if op == "+":
        newacc = acc + int(val)
    elif op == '-':
        newacc = acc - int(val)
    else:
        print ('accerror')
        print (value)
    return newacc
updateacc('-29', 1)
```

```python
def doinc(value, item):
    op = value[:1]
    val = value[1:]
    if op == "+":
        newitem = item + int(val)
    elif op == '-':
        newitem = item - int(val)
    else:
        print ('keyerror')
        print (value)
    return newitem
doinc('+274', 6)
```

```python
def run(org_commands, startkey):
    key = startkey
    acc = 0
    rancommands = {}
    commands = org_commands.copy()
    while key < len(commands):        
        rancommands[key] = commands[key]
        ops = commands[key]
        
        if (ops.rsplit(' ',1)[1] == 'done'):
            return acc, rancommands
        commands[key] = commands[key] + ' done'
        
        opslist = ops.split(' ')
        cmd = opslist[0]
        value = opslist [1]
        if cmd == 'acc':
            acc = updateacc(value, acc)
#         elif cmd == 'nop':
#             #do nothing
        elif cmd == 'jmp':
            key = doinc(value, key)
            continue

        key = key + 1
        
    return acc, rancommands
```

```python
acc, commands = run(dict_data,0)
```

```python
acc, commands = run(dict_testdata,0)
```

```python
def testchange(key, data):
    commands = data.copy()
    commands[key] = 'nop +0'
    acc, newcommands = run(commands,0)
    return acc, newcommands
testchange(265, dict_data)
```
